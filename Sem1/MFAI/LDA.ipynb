```python
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt

# # Example data
# data = {
#     'Result': ['Passed', 'Passed', 'Passed', 'Passed', 'Passed', 'Passed', 'Failed', 'Failed', 'Failed', 'Failed', 'Failed', 'Failed' ],
#     'Curvature': [40, 11.1, 30, 21.4, 10.7, 3.4, 42, 31.1, 50, 60.4, 45.7, 17.3],
#     'Diameter': [36, 37.2, 36.5, 39.4, 39.6, 40.7, 37.6, 42.2, 38.5, 39.4, 38.6, 42.7]
# }

# data = {
#     'Result': ['Passed', 'Passed', 'Passed', 'Passed', 'Failed', 'Failed', 'Failed'],
#     'Curvature': [2.95, 2.53, 3.57, 3.16, 2.58, 2.16, 3.27],
#     'Diameter': [6.63, 7.79, 5.65, 5.47, 4.46, 6.22, 3.52]
# }

data = {
    'Result': ['Passed', 'Passed', 'Passed', 'Passed', 'Passed', 'Passed', 'Passed', 'Passed', 'Failed', 'Failed', 'Failed', 'Failed', 'Failed', 'Failed', 'Failed'],
    'Curvature': [42, 10.3, 33.6, 20.8, 11.9, 2.9, 1.9, 10.4, 42, 31.1, 50, 61.4, 44.7, 17.3, 51],
    'Diameter': [36, 37.2, 36.5, 39.4, 39.6, 40.7, 41.9, 36.3, 37.6, 42.2, 38.5, 38.2, 37.6, 42.7, 38.1]
}

# Create DataFrame
df = pd.DataFrame(data)

# Splitting passed and failed data for separate group statistics
passed = df[df['Result'] == 'Passed']
failed = df[df['Result'] == 'Failed']

# 1. Means
mean_passed_curvature = passed['Curvature'].mean()
mean_passed_diameter = passed['Diameter'].mean()
mean_failed_curvature = failed['Curvature'].mean()
mean_failed_diameter = failed['Diameter'].mean()

mean_total_curvature = df['Curvature'].mean()
mean_total_diameter = df['Diameter'].mean()

# 2. Variances
var_passed_curvature = passed['Curvature'].var()
var_passed_diameter = passed['Diameter'].var()
var_failed_curvature = failed['Curvature'].var()
var_failed_diameter = failed['Diameter'].var()

var_total_curvature = df['Curvature'].var()
var_total_diameter = df['Diameter'].var()

# 3. Covariances
cov_passed = np.cov(passed['Curvature'], passed['Diameter'])[0][1]
cov_failed = np.cov(failed['Curvature'], failed['Diameter'])[0][1]
cov_total = np.cov(df['Curvature'], df['Diameter'])[0][1]

# 4. Between-group matrix
mean_vector_passed = np.array([mean_passed_curvature, mean_passed_diameter])
mean_vector_failed = np.array([mean_failed_curvature, mean_failed_diameter])
overall_mean_vector = np.array([mean_total_curvature, mean_total_diameter])

n_passed = len(passed)
n_failed = len(failed)

# # Between-group matrix (S_B)
# S_B = n_passed * np.outer(mean_vector_passed - overall_mean_vector, mean_vector_passed - overall_mean_vector) + \
#       n_failed * np.outer(mean_vector_failed - overall_mean_vector, mean_vector_failed - overall_mean_vector)

# 5. Within-group matrix
S_W_passed = np.cov(passed['Curvature'], passed['Diameter']) * (n_passed - 1)
S_W_failed = np.cov(failed['Curvature'], failed['Diameter']) * (n_failed - 1)

# Total within-group scatter matrix (S_W)
S_W = (S_W_passed + S_W_failed) / (n_passed + n_failed - 2)
T = np.array([[var_total_curvature, cov_total], [cov_total, var_total_diameter]])
S_B = T - S_W
# 6. Eigenvalues and Eigenvectors
eig_vals, eig_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))
S_final = np.linalg.inv(S_W).dot(S_B)
# eig_vecs = eig_vecs * eig_vals

# Sort eigenvalues and corresponding eigenvectors
sorted_indices = np.argsort(eig_vals)[::-1]
eig_vals = eig_vals[sorted_indices]
eig_vecs = eig_vecs[:, sorted_indices]

# 7. Final equation (linear discriminant)
final_equation = eig_vecs[:, 0]

# Output main statistics
print(f"Mean of Passed Curvature: {mean_passed_curvature}")
print(f"Mean of Passed Diameter: {mean_passed_diameter}")
print(f"Mean of Failed Curvature: {mean_failed_curvature}")
print(f"Mean of Failed Diameter: {mean_failed_diameter}")
print(f"Total Passed: {n_passed}")
print(f"Total Failed: {n_failed}")

print(f"Variance of Passed Curvature: {var_passed_curvature}")
print(f"Variance of Passed Diameter: {var_passed_diameter}")
print(f"Variance of Failed Curvature: {var_failed_curvature}")
print(f"Variance of Failed Diameter: {var_failed_diameter}")
print(f"Variance of Total Curvature: {var_total_curvature}")
print(f"Variance of Total Diameter: {var_total_diameter}")

print(f"Covariance of Passed (Curvature, Diameter): {cov_passed}")
print(f"Covariance of Failed (Curvature, Diameter): {cov_failed}")
print(f"Covariance of Total (Curvature, Diameter): {cov_total}")

print(f"Between-group Matrix (S_B): \n{S_B}")
print(f"Within-group Matrix (S_W): \n{S_W}")
print(f"S Matrix (S): \n{S_final}")

print(f"Eigenvalues: {eig_vals}")
print(f"Eigenvectors: \n{eig_vecs}")

print(f"Final LDA Equation Coefficients: {final_equation}")

# Initialize lists for the extended table
C_group_mean = []
C_prime = []
D_group_mean = []
D_prime = []
C_prime_squared = []
D_prime_squared = []
C_prime_D_prime = []

C_total_mean = [mean_total_curvature] * len(df)
C_double_prime = []
D_total_mean = [mean_total_diameter] * len(df)
D_double_prime = []
C_double_prime_squared = []
D_double_prime_squared = []
C_double_prime_D_double_prime = []

# Calculate all values row by row
for index, row in df.iterrows():
    # Group-wise means
    if row['Result'] == 'Passed':
        c_mean = mean_passed_curvature
        d_mean = mean_passed_diameter
    else:
        c_mean = mean_failed_curvature
        d_mean = mean_failed_diameter

    C_group_mean.append(c_mean)
    D_group_mean.append(d_mean)

    # C' and D' (C - C_group_mean, D - D_group_mean)
    c_prime = row['Curvature'] - c_mean
    d_prime = row['Diameter'] - d_mean
    C_prime.append(c_prime)
    D_prime.append(d_prime)

    # C'², D'², C' * D'
    C_prime_squared.append(c_prime ** 2)
    D_prime_squared.append(d_prime ** 2)
    C_prime_D_prime.append(c_prime * d_prime)

    # C'' and D'' (C - C_total_mean, D - D_total_mean)
    c_double_prime = row['Curvature'] - mean_total_curvature
    d_double_prime = row['Diameter'] - mean_total_diameter
    C_double_prime.append(c_double_prime)
    D_double_prime.append(d_double_prime)

    # C''², D''², C'' * D''
    C_double_prime_squared.append(c_double_prime ** 2)
    D_double_prime_squared.append(d_double_prime ** 2)
    C_double_prime_D_double_prime.append(c_double_prime * d_double_prime)

# 5. LDA scores for both eigenvectors
# First eigenvector (LD1)
w1_1, w2_1 = eig_vecs[:, 0]  # First set of eigenvector coefficients

# Second eigenvector (LD2)
w1_2, w2_2 = eig_vecs[:, 1]  # Second set of eigenvector coefficients

# Calculate LDA Scores
df['LDA_Score_Eigen1'] = df['Curvature'] * w1_1 + df['Diameter'] * w2_1
df['LDA_Score_Eigen2'] = df['Curvature'] * w1_2 + df['Diameter'] * w2_2

# Calculate LDA Scores
df['LDA_Score_Eigen1'] = df['Curvature'] * w1_1 + df['Diameter'] * w2_1
df['LDA_Score_Eigen2'] = df['Curvature'] * w1_2 + df['Diameter'] * w2_2

# 6. Plotting the LDA scores for both eigenvectors
plt.figure(figsize=(8, 6))

# Define x-axis positions for LD1 and LD2
x_ld1 = ['LD1'] * len(df)
x_ld2 = ['LD2'] * len(df)

# Plot Passed scores for LD1 (dots)
for index, row in df.iterrows():
    if row['Result'] == 'Passed':
        plt.scatter('LD1', row['LDA_Score_Eigen1'], color='blue', marker='o', label='Passed' if 'Passed' not in plt.gca().get_legend_handles_labels()[1] else "")
        plt.scatter('LD2', row['LDA_Score_Eigen2'], color='blue', marker='o')

# Plot Failed scores for LD1 (crosses)
for index, row in df.iterrows():
    if row['Result'] == 'Failed':
        plt.scatter('LD1', row['LDA_Score_Eigen1'], color='red', marker='x', label='Failed' if 'Failed' not in plt.gca().get_legend_handles_labels()[1] else "")
        plt.scatter('LD2', row['LDA_Score_Eigen2'], color='red', marker='x')

# Add labels and title
plt.xlabel('Eigenvectors')
plt.ylabel('LDA Score')
plt.title('Comparison of LDA Scores for LD1 and LD2')

# Add a legend
plt.legend()

# Display the plot
plt.show()

# Create the result DataFrame with the requested columns
extended_table = pd.DataFrame({
    'Quality Check Result': df['Result'],
    'Curvature (C)': df['Curvature'],
    'Diameter (D)': df['Diameter'],
    'C_group_mean': C_group_mean,
    'C - C_group_mean (C\')': C_prime,
    'D_group_mean': D_group_mean,
    'D - D_group_mean (D\')': D_prime,
    '(C\') ** 2': C_prime_squared,
    '(D\') ** 2': D_prime_squared,
    'C\' * D\'': C_prime_D_prime,
    'C_total_mean': C_total_mean,
    'C - C_total_mean (C\'\')': C_double_prime,
    'D_total_mean': D_total_mean,
    'D - D_total_mean (D\'\')': D_double_prime,
    '(C\'\') ** 2': C_double_prime_squared,
    '(D\'\') ** 2': D_double_prime_squared,
    'C\'\' * D\'\'': C_double_prime_D_double_prime,
    'LDA_Score_Eigen1': df['LDA_Score_Eigen1'],  # Adding LD1 scores
    'LDA_Score_Eigen2': df['LDA_Score_Eigen2']   # Adding LD2 scores
})

# Print the extended table
print("\nExtended Table:\n")

with pd.option_context('display.max_columns', 500):
    display(extended_table)
```
