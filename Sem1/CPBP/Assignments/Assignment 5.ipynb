```python
import requests

def fetch_url_data(url):
    try:
        # Send a GET request to the specified URL
        response = requests.get(url)
        # Check if the request was successful
        response.raise_for_status()
        # Print the content of the response
        print("Data from URL:")
        print(response.text)
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data from URL: {e}")

# Example usage
url = "https://shuchikas.github.io/ccet-projects/dma_project/index.html"  # Replace with your URL
fetch_url_data(url)
```

```python
# Write a program that counts the number of tabs, spaces and newline characters in a file.

t = 0
s = 0
n = 0
with open("C:/Users/shuch/NITJ/Sem1/CPBP/Class Work/file_handling_test/file1.txt") as f1:
    a = f1.read()
    for i in a:
        if i == '\t':
            t += 1
        elif i == ' ': 
            s += 1
        elif i == '\n':
            n += 1
print("Tabs: ", t)
print("Spaces: ", s)
print("Newline: ", n)
```

```python
# Write a program that reads data from the file and calculates the percentage of vowels and consonants in the file.

vowels = "aeiouAEIOU"
v = 0
c = 0
with open("C:/Users/shuch/NITJ/Sem1/CPBP/Class Work/file_handling_test/file1.txt", "r") as f1:
    a = f1.read()
    for ch in a:
        if ch.isalpha() and ch in vowels:
            v+=1
        elif ch.isalpha() and ch not in vowels:
            c+=1
print("Consonants = ", c)
print("Vowels = ", v)

print("Percentage of consonants: ", c/(c+v)*100)
print("Percentage of vowels: ", v/(c+v)*100)
```

```python
# What are different access modes in which you can open the file.

In Python, files can be opened using different access modes that define how the file should be opened (read, write, append, etc.). 
These modes are specified as a string when calling the open() function. Below are the commonly used file access modes:

1. Read Modes
    'r': Read-only mode (default mode)
        Opens the file for reading.
        File must exist, or an error is raised.
        Example: open('file.txt', 'r')
    'rb': Read-only mode (binary)
        Opens the file for reading in binary format.
        Used for non-text files like images or PDFs.
        Example: open('file.bin', 'rb')
2. Write Modes
    'w': Write-only mode
        Opens the file for writing.
        If the file exists, its contents are overwritten.
        If the file does not exist, it is created.
        Example: open('file.txt', 'w')
    'wb': Write-only mode (binary) 
        Opens the file for writing in binary format.
        Example: open('file.bin', 'wb')
3. Append Modes
    'a': Append mode
        Opens the file for writing, appending new data to the end of the file.
        If the file does not exist, it is created.
        Example: open('file.txt', 'a')
    'ab': Append mode (binary)
        Opens the file for appending in binary format.
        Example: open('file.bin', 'ab')
4. Read and Write Modes
    'r+': Read and write mode
        Opens the file for both reading and writing.
        File must exist, or an error is raised.
        Example: open('file.txt', 'r+')
    'rb+': Read and write mode (binary)
        Opens the file for both reading and writing in binary format.
        Example: open('file.bin', 'rb+')
    'w+': Write and read mode
        Opens the file for both writing and reading.
        If the file exists, its contents are overwritten.
        If the file does not exist, it is created.
        Example: open('file.txt', 'w+')
    'wb+': Write and read mode (binary)
        Opens the file for both writing and reading in binary format.
        Example: open('file.bin', 'wb+')
    'a+': Append and read mode
        Opens the file for both appending and reading.
        If the file does not exist, it is created.
        Example: open('file.txt', 'a+')
    'ab+': Append and read mode (binary)
        Opens the file for both appending and reading in binary format.
        Example: open('file.bin', 'ab+')
```

```python
# Write a program that writes data to a file in such a way that each character after a full stop is capitalized and all the numbers are written in brackets.

import re
rand_text = "this is an example text. it includes some sentences with numbers like 123 and 456. also, it should capitalize letters after full stops. test it with some random sentences. another number is789"
print("-----Input String: -----\n\n", rand_text)
pat = r"\. (\w)"
a = re.finditer(pat, rand_text)
for mat in a:
    x = mat.start()
    y = mat.end()
    rep = mat.group(1).upper()
    rand_text = rand_text[:x + 2] + rep + rand_text[y:]

pat2 = r'(\d+)'
rand_text = re.sub(pat2, r'{\1}', rand_text)

f2 = open(r"C:\Users\shuch\NITJ\Sem1\CPBP\Class Work\file_handling_test\file2.txt"  , "a+")
f2.write(rand_text)
f2.seek(0)
print("\n\n-----New contents of file2: -----\n\n", f2.read(), "\n\n")

f2.close()
```
