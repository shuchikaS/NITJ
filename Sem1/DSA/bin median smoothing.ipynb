```python
import numpy as np

# Dataset
data = [5, 13, 11, 10, 35, 15, 55, 72, 50, 215, 204, 92]
sorted_data = sorted(data)

# Number of bins
n_bins = 4

# Split data into bins
bins = np.array_split(sorted_data, n_bins)

# Perform median smoothing
smoothed_data = []
for b in bins:
    median = np.median(b)
    smoothed_data.extend([median] * len(b))

# Print original and smoothed data
print("Original Data: ", sorted_data)
print("Smoothed Data: ", smoothed_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
data = [5, 13, 11, 10, 35, 15, 55, 72, 50, 215, 204, 92]
sorted_data = sorted(data)

# Number of bins
n_bins = 4

# Split data into bins
bins = np.array_split(sorted_data, n_bins)

# Perform median smoothing
smoothed_data = []
for b in bins:
    median = np.median(b)
    smoothed_data.extend([median] * len(b))

# Plot original vs smoothed data histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Original Data Histogram
ax1.hist(sorted_data, bins=n_bins, edgecolor='black', rwidth=0.8)
ax1.set_title('Original Data Histogram')
ax1.set_xlabel('Value Ranges')
ax1.set_ylabel('Frequency')

# Smoothed Data Histogram
ax2.hist(smoothed_data, bins=n_bins, edgecolor='black', rwidth=0.8)
ax2.set_title('Smoothed Data Histogram (Bin Median Smoothing)')
ax2.set_xlabel('Value Ranges')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Retrying the plotting process

import numpy as np
import matplotlib.pyplot as plt

# Dataset
data = [5, 13, 11, 10, 35, 15, 55, 72, 50, 215, 204, 92]
sorted_data = sorted(data)

# Equal Frequency
n_bins_freq = 4  # Number of bins for equal frequency
data_per_bin = len(sorted_data) // n_bins_freq

# Create equal frequency bins
freq_bins = [sorted_data[i * data_per_bin:(i + 1) * data_per_bin] for i in range(n_bins_freq)]
freq_bins_ranges = [(bin[0], bin[-1]) for bin in freq_bins]

# Equal Width
n_bins_width = 4  # Number of bins for equal width
bin_width = (max(data) - min(data)) / n_bins_width
width_bins = np.arange(min(data), max(data) + bin_width, bin_width)

# Plot histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Equal Frequency Histogram
ax1.hist(sorted_data, bins=[5, 11, 35, 72, 215], edgecolor='black', rwidth=0.8)
ax1.set_title('Equal Frequency Histogram')
ax1.set_xlabel('Value Ranges')
ax1.set_ylabel('Frequency')
ax1.set_xticks([5, 11, 35, 72, 215])
ax1.set_xticklabels(['5-11', '13-35', '50-72', '92-215'])

# Equal Width Histogram
ax2.hist(data, bins=width_bins, edgecolor='black', rwidth=0.8)
ax2.set_title('Equal Width Histogram')
ax2.set_xlabel('Value Ranges')
ax2.set_ylabel('Frequency')
ax2.set_xticks(width_bins)
ax2.set_xticklabels([f'{int(b)}-{int(b + bin_width)}' for b in width_bins[:-1]])

plt.tight_layout()
plt.show()
```
