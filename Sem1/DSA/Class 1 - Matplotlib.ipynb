```python
days = range(1,9)
cel = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
fig, ax = plt.subplots()
ax.plot(days, cel)
ax.set(xlabel = 'Day',
      ylabel = 'Temp in Celcius', 
      title = 'Temp Graph')
plt.show()
```

```python
# Multiple plots in one graph

days = list(range(1, 5))
cel_min = [19.6, 24.1, 26.7, 28.3]
cel_max = [25.6, 25.1, 28.7, 48.3]
fig, ax = plt.subplots()
ax.plot(days, cel_min,
        days, cel_min, "py",
        days, cel_max,
        days, cel_max, "pr"
       )
ax.set(xlabel = 'Day',
      ylabel = 'Temp in Celcius', 
      title = 'Temp Graph')
plt.show()
```

```python
# Multiple plots in one graph

days = list(range(1, 5))
cel_min = [19.6, 24.1, 26.7, 28.3]
cel_max = [25.6, 25.1, 28.7, 48.3]
fig, ax = plt.subplots()
ax.plot(days, cel_min)
ax.plot(days, cel_min, "py")
ax.plot(days, cel_max)
ax.plot(days, cel_max, "pr")
ax.set(xlabel = 'Day',
      ylabel = 'Temp in Celcius', 
      title = 'Temp Graph')
plt.show()
```

```python
# Bar chart example

import numpy as np
years = [str(year) for year in range(2010, 2021)]
visitors = (1241, 50927, 162242, 222093,
           665004, 2071987, 2160407, 3789215,
           5399000, 5474016, 6003672)
plt.bar(years, visitors, color = "green")
plt.xlabel("Years")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.plot()
plt.show()
```

```python
# Histogram

gaussian_numbers = [np.random.normal(size = 10000)]
plt.hist(gaussian_numbers, bins = 20)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.plot()
plt.show()
```

```python
# Scatter Plots

x = np.arange(0, 11)

y1 = np.random.randint(2, 7, (11,))
y2 = np.random.randint(9, 14, (11,))
y3 = np.random.randint(15, 26, (11,))

plt.scatter(x, y1)
plt.scatter(x, y2, marker='v', color='r')
plt.scatter(x, y3, marker='^', color='m')
plt.title('Scatter Plot Example')
plt.show()
```

```python
# Stack plots

idxes = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [23, 42, 33, 43, 8, 14, 43, 18, 21]
y2 = [9, 31, 25, 14, 17, 17, 42, 27, 28]
y3 = [18, 29, 19, 22, 18, 16, 13, 32, 21]
plt.stackplot(idxes, y1, y2, y3)
plt.title('Stack Plot Example')
plt.show()
```

```python
# Pie Charts

labels = 'C', 'Python', 'Java', 'C++', 'C#'
sizes = [13.38, 11.87, 11.74, 7.87, 4.41]
explode = (0, 0.1, 0, 0, 0)

fig1, axl = plt.subplots()
axl.pie(sizes, explode=explode, labels = labels, autopct='%1.1f%%', shadow = True, startangle = 0)
axl.axis('equal')

plt.title('TIOBE Index for May 2021')
plt.show()
```

```python
# Pie Chart where the slices will be ordered and plotted counter-clockwise

labels = 'C', 'Python', 'Java', 'C++', 'C#', 'others'
sizes = [13.38, 11.87, 11.74, 7.87, 4.41]
sizes.append(100 - sum(sizes))
explode = (0, 0.1, 0, 0, 0, 0)

fig1, axl = plt.subplots()
axl.pie(sizes, explode=explode, labels = labels, autopct='%1.1f%%', shadow = True, startangle = 0)
axl.axis('equal')

plt.title('TIOBE Index for May 2021')
plt.show()
```
