```python
# Scaling

import cv2
import numpy as np

img = cv2.imread('img.jpg')

res = cv2.resize(img, None, fx = 2, fy = 2, interpolation = cv2.INTER_CUBIC)

import matplotlib.pyplot as plt

plt.imshow(img)
plt.show()
plt.imshow(res)
plt.show()
```

```python
# Scaling

import cv2
import numpy as np

img = cv2.imread('img.jpg')

res = cv2.resize(img, None, fx = 2, fy = 2, interpolation = cv2.INTER_LINEAR)

import matplotlib.pyplot as plt

plt.imshow(img)
plt.show()
plt.imshow(res)
plt.show()
```

```python
cv2.imshow('img_orig', img)
cv2.waitKey(0)

cv2.imshow('new', res)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# WARP-AFFINE

import cv2
import numpy as np

img = cv2.imread('img.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])

dst = cv2.warpAffine(img, M, (cols,rows))

import matplotlib.pyplot as plt
plt.imshow(dst)
plt.show()
```

```python
cv2.imshow('img', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# ROTATION

import cv2
import numpy as np

img = cv2.imread('img.jpg',0)
rows,cols = img.shape

# cv2.getRotationMatrix2D(center, angle, scale)
M = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1)
dst = cv2.warpAffine(img, M, (cols,rows))
import matplotlib.pyplot as plt
plt.imshow(dst)
plt.show()
```

```python
cv2.imshow('img', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
