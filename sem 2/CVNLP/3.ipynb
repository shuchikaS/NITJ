```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('img.jpg')
image = cv2.resize(image, (300, 300))
b = image[:, :, 0:1]
g = image[:, :, 1:2]
r = image[:, :, 2:3]
# RGB - Blue
b_rgb = cv2.cvtColor(b, cv2.COLOR_BGR2RGB)
plt.subplot(131), plt.imshow(b_rgb)
plt.title('Blue')

# RGB - Green
g_rgb = cv2.cvtColor(g, cv2.COLOR_BGR2RGB)
plt.subplot(132), plt.imshow(g_rgb)
plt.title('Green')

# RGB - Blue
r_rgb = cv2.cvtColor(r, cv2.COLOR_BGR2RGB)
plt.subplot(133), plt.imshow(r_rgb)
plt.title('Red')

plt.tight_layout()
plt.show()

# cv2.imshow('blue', b)
# cv2.imshow('green', g)
# cv2.imshow('red', r)

# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
import numpy as np

img = cv2.imread("img.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
plt.subplot(131), plt.imshow(img_rgb)
plt.title('RGB')

plt.subplot(132), plt.imshow(img)
plt.title('BGR')

plt.subplot(133), plt.imshow(img_hsv)
plt.title('HSV')

plt.tight_layout()
plt.show()
```

```python
def warm_cool_detect():
    img = cv2.imread("img.jpg")
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
    img_hue = img_hsv[:, :, 0]

    cool = (img_hue >= 90) & (img_hue <= 180) 
    warm = (img_hue < 90) & (img_hue >= 0) 

    warm_percent = np.sum(warm) / img_hue.size * 100
    cool_percent = np.sum(cool) / img_hue.size * 100

    if warm_percent > cool_percent:
        print(f"Image is {warm_percent:.2f}% warm")
    else:
        print(f"Image is {cool_percent:.2f}% cool")
warm_cool_detect()
```
