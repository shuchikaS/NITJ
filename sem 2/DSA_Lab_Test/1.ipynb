```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Drop rows with missing values in 'fare', 'embarked', or 'pclass'
titanic = titanic.dropna(subset=['fare', 'embarked', 'pclass'])

# Set plot style
sns.set_style('whitegrid')

# Create boxplot with hue as 'pclass'
plt.figure(figsize=(10, 12))
ax = sns.boxplot(data=titanic, x='embarked', y='fare', hue='pclass')

# Calculate mean fare per embarked group
mean_fares = titanic.groupby('embarked')['fare'].mean()

# Overlay mean fare line
order = ['S', 'C', 'Q']

# Reindex the mean_fares Series to match this order
mean_fares_ordered = mean_fares.reindex(order)

# Now plot
plt.plot(mean_fares_ordered.index, mean_fares_ordered.values, 
         color='black', marker='o', linestyle='-', linewidth=2, label='Mean Fare')
# Add legend for mean line
plt.legend()

# Set labels and title
plt.title('Boxplot of Fare grouped by Embarked with Pclass Hue and Mean Fare Line')
plt.xlabel('Embarked')
plt.ylabel('Fare')

# Show plot
plt.show()
```
