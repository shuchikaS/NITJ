```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# dataframe = pd.read_csv("house-prices-advanced-regression-techniques/train.csv")
dataframe = pd.read_csv("AmesHousing.csv")

df = dataframe.select_dtypes(include=['number'])

df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
plt.figure(figsize=(16, 6))  # Set figure size before plotting
heatmap = sns.heatmap(df.corr())

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(16, 6))


# Create heatmap
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=False, cmap="BrBG")
# heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap="coolwarm")

# Set the title correctly
heatmap.set_title("Correlation Heatmap", fontdict={'fontsize': 18}, pad=12)

# Show the plot
plt.show()
```

```python
plt.savefig('heatmap.png', dpi = 100, bbox_inches = 'tight')
plt.show()
```

```python
# Triangle correlation heatmap

# Cells will be invisible for values where mask is True

np.triu(np.ones_like(df.corr()))
```

```python
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(df.corr(), dtype = np.bool))
heatmap = sns.heatmap(df.corr(), mask = mask, vmin = -1, vmax = 1, annot=False, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize': 18}, pad=16)
```

# Correlation of independent variables with dependent variables

```python
df.corr()[['Order']].sort_values(by='Order', ascending = False)
```

```python
plt.figure(figsize=(16, 6))


# Create heatmap
heatmap = sns.heatmap(df.corr()[['Year Built']].sort_values(by='Year Built', ascending = False), vmin=-1, vmax=1, annot=False, cmap="BrBG")
# heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap="coolwarm")

# Set the title correctly
heatmap.set_title("Features correlating with Year Built", fontdict={'fontsize': 18}, pad=12)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating the dataset
data = {
    "Color": ["Red", "Green", "Red", "Yellow", "Green", "Yellow", "Yellow", "Yellow", "Orange", "Brown",
              "Orange", "Orange", "Yellow", "Green", "Green", "Purple", "Purple", "Purple", "Red", "Purple"],
    "Weight (g)": [150, 140, 160, 145, 155, 120, 110, 130, 200, 900,
                   200, 190, 180, 170, 195, 5, 6, 4, 5, 6],
    "Sweetness (1-10)": [7, 6, 8, 7, 7, 9, 8, 9, 10, 8,
                          6, 7, 4, 5, 7, 8, 9, 7, 8, 9],
    "Acidity (pH)": [3.5, 3.7, 3.6, 3.8, 3.5, 4.5, 4.6, 4.4, 4.1, 3.9,
                      3.0, 3.1, 2.8, 2.9, 3.1, 3.4, 3.3, 3.5, 3.2, 3.1],
    "Class": ["Pome Fruit", "Pome Fruit", "Pome Fruit", "Pome Fruit", "Pome Fruit",
              "Tropical Fruit", "Tropical Fruit", "Tropical Fruit", "Tropical Fruit", "Tropical Fruit",
              "Citrus Fruit", "Citrus Fruit", "Citrus Fruit", "Citrus Fruit", "Citrus Fruit",
              "Berry", "Berry", "Berry", "Berry", "Berry"]
}

# Convert to DataFrame
df = pd.DataFrame(data)
print(df.head(20))

# Encode categorical variables (Class and Color) using label encoding
df["Class"] = df["Class"].astype("category").cat.codes  # Encode Class (Pome=0, Tropical=1, Citrus=2, Berry=3)
df["Color"] = df["Color"].astype("category").cat.codes  # Encode Color

# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="BrBG", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Features")
plt.show()
```
