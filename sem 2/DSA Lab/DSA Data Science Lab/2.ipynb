```python
# CountPlot - Returns count of total values for each category using bars
import pandas as pd
mtcars = pd.read_csv('C:/Users/shuch/NITJ/Sem 2/DSA Data Science Lab/mtcars.csv')

sns.countplot(x = 'cyl', data = mtcars, palette = "Set1")
```

```python
sns.countplot(x = 'carb', data = mtcars, palette = "Set1")
```

```python
sns.countplot(y = 'gear', data = mtcars, palette = "rocket")
```

```python
# Grouped Count
# Below gives count of cars for each category of gears that are grouped based on the number of cylinders

sns.countplot(x = 'gear', hue = 'cyl', data = mtcars, palette = 'Set1')
```

```python
# Distribution Plot
# Plots the dist of miles per gallon of the different vehicles. mpg = total distance the car can travel per gallon of fuel

sns.distplot(mtcars.mpg, bins = 10, color = 'g')
```

```python
# Heatmap
cols = list(pd.read_csv("C:/Users/shuch/NITJ/Sem 2/DSA Data Science Lab/mtcars.csv", nrows=1))
mtcars2 = pd.read_csv('C:/Users/shuch/NITJ/Sem 2/DSA Data Science Lab/mtcars.csv', usecols =[i for i in cols if i != "model"])

sns.heatmap(mtcars2.corr(), cbar = True, linewidths = 0.5)
```
